Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image
dengan tipe data int16.

```python
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

```python
#perhatikan koordinat titik pt1 dan pt2
cv.rectangle(black_img,pt1=(384,0), pt2=(510,150), color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2
yang tertulis pada kode program.

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300), color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang
tertulis pada kode program.

```python
cv.circle(black_img, center=(100,100), radius=50, color=(250,0,0), thickness=8)
plt.imshow(black_img)
```

Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.

```python
cv.line(black_img, pt1=(0,0), pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
ditentukan.

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(10,500), fontFace=font,fontScale=4, color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

Pembuatan black image kembali dilakukan dengan tipe data int32

```python
black_img2=np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
```

Berikut adalah kode program untuk inisialisasi NumPy array dengan tipe data int32

```python
vertices = np.array([[100, 300], [200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

Array tersebut kemudian di reshape sebagai berikut

```python
pts = vertices.reshape((-1,1,2,)) #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yang mewakili R, G, B
pts
```

Penambahan polyline pada black image kedua yang telah dibuat.

```python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```

1. Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib?
Jawab :
Perbedaan antara menampilkan gambar dengan OpenCV (cv2_imshow) dan Matplotlib (plt.imshow) terletak pada cara dan konteks tampilan. OpenCV menampilkan gambar dalam jendela terpisah yang sering digunakan dalam aplikasi desktop dan GUI, sedangkan Matplotlib menampilkan gambar sebagai bagian dari plot di dalam notebook atau lingkungan interaktif. Matplotlib lebih cocok untuk analisis data dan presentasi visual dalam notebook, memberikan kontrol lebih atas tata letak dan integrasi dengan grafik lainnya. Sebaliknya, OpenCV lebih sesuai untuk aplikasi berbasis desktop yang memerlukan antarmuka pengguna grafis.

2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16
dan int32?
Jawab : Pembuatan gambar hitam dengan tipe data int16 dan int32 berbeda dalam hal rentang nilai yang dapat disimpan. int16 memiliki rentang nilai dari -32,768 hingga 32,767, yang cukup untuk gambar hitam tetapi dapat mengalami overflow jika nilai melebihi rentang ini. Sebaliknya, int32 memiliki rentang yang jauh lebih besar, yaitu -2,147,483,648 hingga 2,147,483,647, sehingga lebih aman dari overflow dan memungkinkan penyimpanan nilai yang lebih besar tanpa masalah. Meskipun int32 menggunakan lebih banyak memori, ia memberikan fleksibilitas lebih besar dalam pemrosesan gambar dibandingkan int16.
